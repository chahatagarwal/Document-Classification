```python
import os
import re
import pandas as pd
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

### Get labels of dataset

```python
contents=os.listdir("Data/bbc")
regex = re.compile('[A-Za-z0-9]*-test')
labels = [x for x in contents if not regex.match(x)]
```

```python
labels
```

### Reading .txt files and basic pre-processing to raw text data

Use labels to know folder names 

**Training Data** : the folder with label name contains the training data<br>
**Testing Data** : the folder with label+"-test" extension is containing testing data

**0 - Convert to lower alphabetic form** <br> 
Convert all data to lower alphabetic form for that analysis should be only for one type of letters<br>
The meaning doesn't change with change in letter form

**1 - Remove markup language tags** <br> 
Strip all the HTML or XML tags, if they are in data 

**2 - Remove numbers and seperate words** <br> 
Numbers are use less in analysis of text<br>
strip all the numbers <br>
create a list of words from raw strings

**3 - Remove stop words** <br> 
stop words are words which are too common in text and doesn't convey any meaning when we analyse as bag of words model<br>
example = the, a, was...

**4 - Stemming or Lemmatization** <br> 
This both do same thing as they convert near by meaning word to common word <br>

Example:<br>
![Example of Stemming and Lemmatization](http://basistech.com/wp-content/uploads/2013/12/lemma_example.png)<br>

[Get More Detailse on Stemming and Lemmatization!](https://nlp.stanford.edu/IR-book/html/htmledition/stemming-and-lemmatization-1.html)

```python
File_vs_Label_Training = pd.DataFrame(labels,columns=["Label"])
File_vs_Label_Testing = pd.DataFrame(labels,columns=["Label"])
no_of_training_files = []
no_of_testing_files = []
```

```python
Training_Data=[]
Testing_Data=[]

for label in labels:
    i=0
#-----------------------------------------------------Training Data------------------------------------------------------
    file_list = os.listdir("Data/bbc/"+label)
    for file in file_list:
        
        f=open("Data/bbc/"+label+"/"+file)
        
        data = ""
        lines = f.readlines()
        for line in lines:
            data=data+line.strip()

        f.close()
        
        #Data Filtering
        
        # 0
        data = data.lower()
        
        # 1
        soup = BeautifulSoup(data,"lxml")
        data = soup.get_text()
        
        # 2
        data = re.findall(r"(?i)\b[a-z]+\b", data)
        
        # 3
        stop_words = set(stopwords.words('english'))
        temp=data
        data=[]
        for w in temp:
            if w not in stop_words:
                data.append(w)
                
        # 4
        lemmatiser = WordNetLemmatizer()
        temp=data
        data=[]
        for w in temp:
            data.append(lemmatiser.lemmatize(w))
           
        Training_Data.append([label,data])
        i+=1
    no_of_training_files.append(i)
     
        
    j=0
#-------------------------------------------------------------Testing Data-------------------------------------------
    file_list = os.listdir("Data/bbc/"+label+"-test")
    for file in file_list:
        
        f=open("Data/bbc/"+label+"-test/"+file)
        
        data = ""
        lines = f.readlines()
        for line in lines:
            data=data+line.strip()

        f.close()
        
        # 0
        data = data.lower()
        
        # 1
        soup = BeautifulSoup(data,"lxml")
        data = soup.get_text()
        
        # 2
        data = re.findall(r"(?i)\b[a-z]+\b", data)
        
        # 3
        stop_words = set(stopwords.words('english'))
        temp=data
        data=[]
        for w in temp:
            if w not in stop_words:
                data.append(w)
                
        # 4
        lemmatiser = WordNetLemmatizer()
        temp=data
        data=[]
        for w in temp:
            data.append(lemmatiser.lemmatize(w))
           
        Testing_Data.append([label,data])
        j+=1
    no_of_testing_files.append(j)
```

```python
File_vs_Label_Training["Files Count"] = no_of_training_files
File_vs_Label_Training = File_vs_Label_Training.set_index("Label")
File_vs_Label_Training.to_csv("Data/Training File Count.csv")

File_vs_Label_Testing["Files Count"] = no_of_testing_files
File_vs_Label_Testing = File_vs_Label_Testing.set_index("Label")
File_vs_Label_Testing.to_csv("Data/Testing File Count.csv")
```

### Creating Data Frame from list

```python
headers = ["Label","Words"]

Training_Data_Frame = pd.DataFrame(Training_Data)
Training_Data_Frame.columns = headers

Testing_Data_Frame = pd.DataFrame(Testing_Data)
Testing_Data_Frame.columns = headers
```

### Finding Uniques Words from all data

```python
unique_words=[]

for document in Training_Data_Frame["Words"]:
    for word in document:
        if word not in unique_words:
            unique_words.append(word)
print(unique_words)
print(len(unique_words))
```

### Creating the Word count matrix

```python
Training_Label_List = Training_Data_Frame["Label"]
Training_Word_Count = pd.DataFrame(Training_Label_List)
Training_Word_Count.columns = ["Label"]

Testing_Label_List = Testing_Data_Frame["Label"]
Testing_Word_Count = pd.DataFrame(Testing_Label_List)
Testing_Word_Count.columns = ["Label"]

word_count = len(unique_words)


print("Training File in Processing")

Counter=0
for unique_word in unique_words:
    
    if Counter%2000==0:
        print(str(int(Counter*100/word_count))+"% Complete")
        
    Current_Word_Count_List=[]
    for document in Training_Data_Frame["Words"]:
        Current_Word_Count_List.append(document.count(unique_word))
    Training_Word_Count[unique_word] = Current_Word_Count_List
    Counter+=1

print("Training File in Process Completed")
print("")   
print("")   
print("Testing File in Processing")

Counter=0
for unique_word in unique_words:
    
    if Counter%2000==0:
        print(str(int(Counter*100/word_count))+"% Complete")
        
    Current_Word_Count_List=[]
    for document in Testing_Data_Frame["Words"]:
        Current_Word_Count_List.append(document.count(unique_word))
    Testing_Word_Count[unique_word] = Current_Word_Count_List
    Counter+=1
    
print("Testing File in Process Completed")
```
