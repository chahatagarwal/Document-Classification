# K-Means

```python
import pandas as pd
import numpy as np
import scipy.spatial.distance as scidis
import random as rand
```

### Un-Normalized data Retrival

```python
Training_Frame = pd.read_csv("Data/TF-IDF Training - Normalized.csv")
Training_Frame = Training_Frame.rename(columns = {'Unnamed: 0':'Label'})
Training_Frame = Training_Frame.set_index("Label")
Training_Frame
```

```python
Training_Data_Labels = list(Training_Frame.index)
Training_Data_Array = np.array(Training_Frame)
```

```python
Labels=[]
for Label in Training_Data_Labels:
    if Label not in Labels:
        Labels.append(Label)
```

```python
no_of_docs=[0]*len(Labels)
i=0
while i<len(no_of_docs):
    no_of_docs[i] += Training_Data_Labels.count(Labels[i])
    i+=1
no_of_docs
```

### Normalized Data Retrival

```python
Training_Frame_Normalized = pd.read_csv("Data/TF-IDF Training + Normalized.csv")
Training_Frame_Normalized = Training_Frame_Normalized.rename(columns = {'Unnamed: 0':'Label'})
Training_Frame_Normalized = Training_Frame_Normalized.set_index("Label")
Training_Frame_Normalized
```

```python
Training_Data_Labels_Normalized = list(Training_Frame_Normalized.index)
Training_Data_Array_Normalized = np.array(Training_Frame_Normalized)
```

```python
Labels_Normalized=[]
for Label in Training_Data_Labels_Normalized:
    if Label not in Labels_Normalized:
        Labels_Normalized.append(Label)
```

```python
no_of_docs_Normalized=[0]*len(Labels_Normalized)
i=0
while i<len(no_of_docs_Normalized):
    no_of_docs_Normalized[i] += Training_Data_Labels_Normalized.count(Labels_Normalized[i])
    i+=1
no_of_docs_Normalized
```

## K-Means Algorithm

```python
z=0
while z<2:
    
#---------------------------------------------Two Type Of Training Sets---------------------------------------------------
    if z==0:
        print("===============================================Normalized Data====================================================")
        Current_Data_Frame = np.array(Training_Data_Array_Normalized)
        Current_Data_Labels = Training_Data_Labels_Normalized
    elif z==1:
        print("=============================================Un-Normalized Data===================================================")
        Current_Data_Frame = np.array(Training_Data_Array)
        Current_Data_Labels = Training_Data_Labels
        
        
        
    np.random.shuffle(Current_Data_Frame)  
    y=0
    while y<4:
        print("")
        if y==0:
            print("-----------------------------Minkwoski - 1-------------------------------")
        elif y==1:
            print("-----------------------------Minkwoski - 3-------------------------------")
        elif y==2:
            print("-------------------------------Eucledian---------------------------------")
        elif y==3:
            print("------------------------------Correlation--------------------------------")


        k=2
        while k<9:
                        
            Assignment_List=[-1]*len(Current_Data_Labels)
            Random_Index_List=[-1]*k

#----------------------------------------------Craete Random Cluster Centres-------------------------------------------
            
            i=0
            while i<k:
                random_no = rand.randint(0,len(Current_Data_Labels))
                if random_no not in Random_Index_List:
                    Random_Index_List[i]=random_no
                    i+=1

            
            Cluster_Mean_Points=[]            
            i=0
            while i<k:
                Point = np.array(Current_Data_Frame[Random_Index_List[i]])
                Cluster_Mean_Points.append(Point)
                i+=1
            Cluster_Mean_Points = np.array(Cluster_Mean_Points)


            Trigger=1
            while Trigger==1:
                Previous_Cluster_Mean_Points = np.array(Cluster_Mean_Points)

                
#----------------------------------------------Find Nearest Cluster Center-------------------------------------------
                i=0
                while i<len(Current_Data_Labels):
                    Min_Distance=0
                    Min_Distance_Cluster=k
                    j=0
                    while j<k:   
                        if y==0:
                            Current_Distance = scidis.minkowski(Cluster_Mean_Points[j],Current_Data_Frame[i],3)
                        elif y==1:
                            Current_Distance = scidis.minkowski(Cluster_Mean_Points[j],Current_Data_Frame[i],3)
                        elif y==2:
                            Current_Distance = scidis.euclidean(Cluster_Mean_Points[j],Current_Data_Frame[i])
                        elif y==3:
                            Current_Distance = scidis.correlation(Cluster_Mean_Points[j],Current_Data_Frame[i])
                        if j==0:
                            Min_Distance = Current_Distance
                            Min_Distance_Cluster = j
                        elif Current_Distance < Min_Distance:
                            Min_Distance = Current_Distance
                            Min_Distance_Cluster = j
                        j+=1
                    Assignment_List[i] = Min_Distance_Cluster
                    i+=1

#----------------------------------------------Craete New Cluster Centres-------------------------------------------
                    
                Cluster_Mean_Points = Cluster_Mean_Points - Cluster_Mean_Points
                i=0
                while i<len(Current_Data_Labels):
                    Cluster_Mean_Points[Assignment_List[i]] += Current_Data_Frame[i]
                    i+=1    
                Assignment_List = list(Assignment_List)
                i=0
                while i<k:
                    Cluster_Mean_Points[i] = Cluster_Mean_Points[i]/Assignment_List.count(i)
                    i+=1

                if (Previous_Cluster_Mean_Points == Cluster_Mean_Points).all():
                    Trigger=0
            print("")
            print("== K : "+str(k)+" ==")
            i=0
            while i<k:
                print(str(i)+"  ---  Spread : "+str(int(Assignment_List.count(i)*100/len(Assignment_List)))+"%  ---  Count : "+str(Assignment_List.count(i)))
                i+=1
            k+=1
        y+=1
    z+=1
```
